##PROJECT TASK07:Which month,year has most animal intakes ?
Draw a pie chart representing year wise animal intake from year 2014 to 2021.

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('DataSet_Animal_Center_Intakes.csv')

# Convert the DateTime column to datetime type
data['DateTime'] = pd.to_datetime(data['DateTime'])

data['Month'] = data['DateTime'].dt.month
data['Year'] = data['DateTime'].dt.year

# Calculate the count of animal intakes per month and year
intake_counts = data.groupby(['Year', 'Month']).size().reset_index(name='Count')

# Find the month and year with the maximum animal intakes
max_intake = intake_counts.loc[intake_counts['Count'].idxmax()]

# Print the month and year with the most animal intakes
print("Month and Year with Most Animal Intakes:")
print("Month:", max_intake['Month'])
print("Year:", max_intake['Year'])

# Filter data from 2014 to 2021
filtered_data = intake_counts[intake_counts['Year'].between(2014, 2021)]

# Calculate the total animal intakes per year
yearly_counts = filtered_data.groupby('Year')['Count'].sum()

# Create a pie chart
plt.pie(yearly_counts, labels=yearly_counts.index, autopct='%1.1f%%')
plt.title('Year-wise Animal Intake (2014-2021)')
plt.axis('equal')
plt.show()
```

##project task08:Create a bar chart to represent month wise animal intake.
Which month has heighest animal intake?

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
data = pd.read_csv('DataSet_Animal_Center_Intakes.csv')

# Convert the DateTime column to datetime type
data['DateTime'] = pd.to_datetime(data['DateTime'])

# Extract the month
data['Month'] = data['DateTime'].dt.month

# Calculate the count of animal intakes per month
intake_counts = data['Month'].value_counts().sort_index()

# Create a bar chart
plt.bar(intake_counts.index, intake_counts.values,color='cyan')
plt.xlabel('Month')
plt.ylabel('Animal Intake Count')
plt.title('Month-wise Animal Intake')
plt.xticks(range(1, 13))
plt.show()

# Find the month with the highest animal intake
max_month = intake_counts.idxmax()

# Print the month with the highest animal intake
print("Month with the Highest Animal Intake:", max_month)
```

##PROJECT TASK09:Create a bar chart to represent animal intake methods wise counting.
What are the animal intake methods ? Print their names?

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('DataSet_Animal_Center_Intakes.csv')
intake_methods = data['Intake Type'].value_counts()

# Print the names of animal intake methods
print("Animal Intake Methods:")
print(intake_methods)
# Calculate the count of animal intakes per intake method
intake_counts = data['Intake Type'].value_counts()

#Create a bar chart
plt.figure(figsize =(15,6))
plt.bar(intake_counts.index,intake_counts.values,color= 'green')
plt.xlabel='intake_counts'
plt.ylabel='count'
plt.title ='animal intake methods wise counting'
plt.xticks(rotation=0)
plt.show()
```
